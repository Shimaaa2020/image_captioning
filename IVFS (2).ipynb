```python
import pandas as pd
data = pd.read_csv('/Users/mehrdad/Downloads/ar-2017-2018-1.csv', header = 0, low_memory=False)
```

```python
from sklearn.preprocessing import OrdinalEncoder

data_enc = OrdinalEncoder()
data["Patient age at treatment"] = data_enc.fit_transform(data[["Patient age at treatment"]])
data["Main reason for producing embroys storing eggs"] = data_enc.fit_transform(data[["Main reason for producing embroys storing eggs"]])
data["Early outcome"] = data_enc.fit_transform(data[["Early outcome"]])
data["Sperm donor age at registration"] = data_enc.fit_transform(data[["Sperm donor age at registration"]])


```

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# run correlation matrix and plot
f, ax = plt.subplots(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool),
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```
